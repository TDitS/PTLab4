```python
import numpy as np
size = 85
# Матрица M
M = np.ones((size, size)) * 2 + np.diag(np.arange(1, 170, 2))
print('Матрица M:')
print(M)
```

```python
# Матрица N
N = np.ones((size, size)) * 4
N[::2, 2::3] = 5
N[1::2, 1::2] = 7
print('Матрица N:')
print(N)
```

```python
# p = {1 + pi: pi ∼ N (2, 4)}^84, i=0: {0, ..., 84}
mx = 2
variance = 4
p = 1 + np.random.normal(mx, variance, size=size)
print('Вектор p:')
print(p)
```

```python
# q^T = {N24,j ⊙ N71,j}^84, j=0: {0, ..., 84}
qT = np.multiply(N[24, :size],N[71, :size])
print('Вектор q^T:')
print(qT)

q = qT.T
print('\nВектор q:')
print(q)
```

```python
# Вычисление f = det(qq^T)p
# Вычисляем произведение qq^T
qqT = np.outer(q, qT)
print('qqT:')
print(qqT)

# Вычисляем определитель матрицы det(qq^T)
det_qqT = np.linalg.det(qqT)
print('\ndet_qqT:')
print(det_qqT)

# Вычисляем f
f = det_qqT * p

# Выводим результат
print('\nВектора f:')
print(f)

f = det_qqT
print('\nВектор f:')
print(f)
```

```python
# Вычисление g = ||{M_(ii) - N_(ii) + q^Tq}^84 ||_2 , где i = 0: {0, ..., 84}
# Создаем диагональную матрицу Mii
Mii = np.diag(M)
print('Mii:')
print(Mii)

# Создаем диагональную матрицу Nii
Nii = np.diag(N)
print('\nNii:')
print(Nii)

# Считаем q^Tq
qTq = np.multiply(qT, q)
print('\nqTq:')
print(qTq)

# Считаем разность Mii - Nii
diff = Mii - Nii
print('\ndiff:')
print(diff)

# Вычисляем {Mii-Nii+q^Tq}
result = diff + qTq
print('\nresult:')
print(result)

# Вычисляем g как норму вектора result
g = np.linalg.norm(result, ord=2) # ||...||2 - нормаль вектора
print('\nВектор g:')
print(g)
```
